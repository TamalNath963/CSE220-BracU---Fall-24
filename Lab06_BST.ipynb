```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
import numpy as np
def find_Path(root, key):
    cap=0
    temp = root
    while temp!=None:
        cap+=1
        if temp.elem==key:
            break
        if temp.elem<key:
            temp = temp.right
        else :
            temp = temp.left

    if temp==None:
        return "No Path Found"
    arr = np.zeros(cap,dtype="int")
    temp = root
    i = 0
    while temp!=None:
        arr[i]=temp.elem
        i+=1
        if temp.elem==key:
            break
        if temp.elem>key:
            temp = temp.left
        else :
            temp = temp.right
    return arr


#DRIVER CODE
#Write by yourself from the given tree

root = BTNode(30)
root.left = BTNode(10)
root.right = BTNode(40)
root.left.left = BTNode(3)
root.right.left = BTNode(35)
root.left.right = BTNode(15)
root.right.right = BTNode(55)


print(find_Path(root,15))
#This should print [30,10,15]

print(find_Path(root,50))
#This should print No Path Found
```

```python
def sum_of_leaves(root, sum):
    #To Do
    if root == None:
      return 0
    else:
      sum = sum_of_leaves(root.left, sum)
      sum += sum_of_leaves(root.right, sum)
    if sum == 0 :
        return root.elem
    return sum

root = BTNode(30)
root.left = BTNode(10)
root.right = BTNode(40)
root.right.left = BTNode(35)
root.right.right = BTNode(55)
root.left.left = BTNode(3)
root.left.right = BTNode(15)
root.left.left.left = BTNode(2)
root.right.left.right = BTNode(36)
#DRIVER CODE
#Write by yourself from the given tree

print(sum_of_leaves(root, 0))
```

```python
def inorder_predecessor(root, x):
  #To Do
  def rightmost(node):
    while node.right :
      node = node.right
    return node

  if root.elem == x.elem and root.left:
    return rightmost(root.left)
  pre = None
  temp = root
  while temp:
    if x.elem > temp.elem:
      pre = temp
      temp = temp.right
    elif x.elem < temp.elem:
      temp = temp.left
    else:
      return pre
  return pre

#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
x = root.left.right.right
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 14: 12
x = root.left.right.left
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 10: 8
```
