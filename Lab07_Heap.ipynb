```python
#Task 01
class MinHeap:
    def __init__(self, cap=10):
        self.__capacity = cap+1
        self.__heap = np.zeros((cap+1),dtype=object)
        self.__size = 1

    def insert(self, value):
        if self.__size == self.__capacity:
            return
        self.__heap[self.__size] = value
        self.swim(self.__size)
        self.__size += 1

    def swim(self, index):
        while index > 1:
            parent = index // 2
            if self.__heap[index] < self.__heap[parent]:
                self.__heap[index], self.__heap[parent] = self.__heap[parent], self.__heap[index]
                index = parent
            else:
                break
    def extractMin(self):
        if self.__size == 1:
            return None
        return self.__heap[1]

    def delete(self):
        if self.__size == 1:
            return None
        minVal = self.__heap[1]
        self.__size -= 1
        self.__heap[1] = self.__heap[self.__size]
        self.sink(0)
        return minVal

    def sink(self, index):
        while 2 * index < self.__size:
            left = 2 * index
            right = 2 * index + 1
            smallest = left

            if right < self.__size and self.__heap[right] < self.__heap[left]:
                smallest = right

            if self.__heap[index] > self.__heap[smallest]:
                self.__heap[index], self.__heap[smallest] = self.__heap[smallest], self.__heap[index]
                index = smallest
            else:
                break

    def sort(self):
        original_size = self.__size
        for i in range(self.__size - 1, 0, -1):
            self.__heap[0], self.__heap[i] = self.__heap[i], self.__heap[0]
            self.__size -= 1
            self.sink(0)
        self.__size = original_size
```

```python
h = MinHeap(7)

h.insert(99)
h.insert(90)
h.insert(40)
h.insert(30)
h.insert(3)
h.insert(7)
h.insert(5)


for i in range(7):
    print(h.extractMin())
    h.delete()
```

```python
#Task 02
class MaxHeap:
    def __init__(self, cap):
        self.__capacity = cap+1
        self.__heap = np.zeros((cap+1),dtype=object)
        self.__size = 1

    def insert(self, value):
        if self.__size == self.__capacity:
            return
        self.__heap[self.__size] = value
        self.swim(self.__size)
        self.__size += 1

    def swim(self, index):
        while index > 1:
            parent = index // 2
            if self.__heap[index] > self.__heap[parent]:
                self.__heap[index], self.__heap[parent] = self.__heap[parent], self.__heap[index]
                index = parent
            else:
                break

    def extractMax(self):
        if self.__size == 1:
            return None
        return self.__heap[1]

    def delete(self):
        if self.__size == 1:
            return None
        maxVal = self.__heap[1]
        self.__size -= 1
        self.__heap[1] = self.__heap[self.__size]
        self.sink(1)
        return maxVal

    def sink(self, index):
        while 2 * index < self.__size:
            left = 2 * index
            right = 2 * index + 1
            largest = left

            if right < self.__size and self.__heap[right] > self.__heap[left]:
                largest = right

            if self.__heap[index] < self.__heap[largest]:
                self.__heap[index], self.__heap[largest] = self.__heap[largest], self.__heap[index]
                index = largest
            else:
                break

    def sort(self):
        original_size = self.__size
        for i in range(self.__size - 1, 0, -1):
            self.__heap[0], self.__heap[i] = self.__heap[i], self.__heap[0]
            self.__size -= 1
            self.sink(0)
        self.__size = original_size
```

```python
h = MaxHeap(7)

h.insert(99)
h.insert(90)
h.insert(40)
h.insert(30)
h.insert(3)
h.insert(7)
h.insert(5)

for i in range(7):
    print(h.extractMax())
    h.delete()
```

```python
#Task 03
def distribute_tasks(tasks, m):
    loads = [0] * m
    for task in tasks:
        min_index = 0
        for i in range(1, m):
            if loads[i] < loads[min_index]:
                min_index = i
        loads[min_index] += task
    return loads

tasks = [2, 4, 7, 1, 6]
m = 4

result = distribute_tasks(tasks, m)
print(result)
```

```python
#Task 04
def top_k_largest(nums, k):
    heap = MaxHeap(len(nums))
    for num in nums:
        heap.insert(num)
    result = []
    for i in range(k):
        result.append(heap.extractMax())
        heap.delete()
    return result
nums = [4, 10, 2, 8, 6, 7]
k = 3
result = top_k_largest(nums, k)
print("Top k Largest Elements:", result)
```
